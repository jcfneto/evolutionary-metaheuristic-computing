# Genetic Algorithm

```python
import numpy as np
from metaevolv.ga import GeneticAlgorithm, GAConfig
```

```python
def rastrigin_function(x: np.ndarray) -> float:
    A = 10
    n = len(x)
    return A * n + np.sum(x**2 - A * np.cos(2 * np.pi * x))
```

```python
config = GAConfig(
    bits=10,
    dimensions=2,
    n_population=200,
    search_range=(-10.0, 15.0),
    k=3,
    cp=0.8,
    mp=0.01,
    max_iter=500,
    selection_type='tournament',
    crossover_type='one_point',
    mutation_type='bit_by_bit',
    pc_variation='constant',
    cp_final=0.9
)
```

```python
ga = GeneticAlgorithm(config=config, obj_function=rastrigin_function)
ga.fit()
```

# Artificial Bee Colony

```python
from metaevolv.abc import ABC, ABCConfig
```

```python
config = ABCConfig(
    search_range=(-10.0, 15.0),
    n_population=200,
    dimensions=2,
    max_iter=50,
    limit=5
)
```

```python
abc = ABC(config=config, obj_function=rastrigin_function)
abc.fit()
```

```python
abc.plot(avg=True)
```
